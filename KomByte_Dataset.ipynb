```python
import csv
import random
from datetime import datetime

# Define the range of dates
start_date = datetime(2018, 1, 1)
end_date = datetime(2023, 12, 31)

# Generate 20,000 random dates
random_dates = [start_date + (end_date - start_date) * random.random() for _ in range(20000)]

# Convert dates to text format
date_strings = [date.strftime('%Y-%m-%d') for date in random_dates]

# Write dates to CSV file
with open('random_dates.csv', 'w', newline='') as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(['Date'])
    for date_str in date_strings:
        writer.writerow([date_str])

print("Random dates have been saved to random_dates.csv")
```
